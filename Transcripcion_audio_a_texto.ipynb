PARTICIONANDO EL AUDIO EN PARTES DE 2 MINUTOS

```python
from pydub import AudioSegment

##Siempre llamar al archivo "test.wav"
# Abrir el archivo de audio
audio = AudioSegment.from_wav("test.wav")

# Calcular la duración del archivo de audio en milisegundos
duration = len(audio)

# Calcular la duración de cada parte en milisegundos (2 minutos = 120000 ms)
part_duration = 120000

# Calcular el número de partes
num_parts = duration // part_duration

# Particionar el archivo
audio_particionado = []
for i in range(num_parts+1):
    start = i * part_duration
    end = (i+1) * part_duration
    part = audio[start:end]
    part.export(f"part{i}.wav", format="wav")
    audio_particionado.append(part.export(f"part{i}.wav", format="wav"))
```

```python
audio_particionado[0].name
```

COLOCANDO LA TRANSCRIPCION EN UN ARCHIVO TXT, CONCATENADO

```python
# Creamos un archivo de texto vacío para guardar el resultado
archivo_total = open("resultado.txt", "w")
r = sr.Recognizer()
    
# Procesamos cada parte del archivo de audio
for i in audio_particionado:
    # Abrimos el archivo de audio
    part_file = i.name
    audio = sr.AudioFile(part_file)

    # Realizar la grabación
    with audio as source:
        audio_data = r.record(source)

    # Convertimos el audio a texto para cada intervalo
    text = r.recognize_google(audio_data, language='es-ES')
    #print(text)

    # Guardamos el texto en el archivo
    archivo_total.write(text + "\n")
print(archivo_total)
```
